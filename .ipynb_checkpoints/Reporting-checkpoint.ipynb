```python
import alpaca_trade_api as tradeapi
import yfinance as yf
import pandas as pd
import numpy as np
import requests
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns

# Alpaca API credentials
API_KEY = 'PKXEO7CC6791U21253KJ'
API_SECRET = 'coJmpQvTefdecGm48U8STtMwDvCEcUnD8Iws3Ghi'
BASE_URL = 'https://paper-api.alpaca.markets'
NEWS_API_KEY = 'fb4ef009a8914047a0c1b1b868800438'

# Initialize Alpaca API
api = tradeapi.REST(API_KEY, API_SECRET, BASE_URL, api_version='v2')

# Sentiment Analyzer
analyzer = SentimentIntensityAnalyzer()

tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA']

# Fetch all activities from the Alpaca account
activities = api.get_activities()

# Print the activities to see the available fields
for activity in activities:
    print(activity)  # or you can use activity._raw to see the raw JSON data

# Fetch account details
account = api.get_account()
portfolio_value = float(account.equity)
today_pnl = float(account.equity) - float(account.last_equity)
print(f"Account equity: ${portfolio_value}")
print(f"Today's PnL: ${today_pnl:.2f}")

# Function to fetch news sentiment using News API
def get_news_sentiment(ticker):
    print(f"Fetching news sentiment for {ticker}")
    url = f'https://newsapi.org/v2/everything?q={ticker}&apiKey={NEWS_API_KEY}'
    response = requests.get(url)
    news_data = response.json()
    if news_data.get('status') == 'ok':
        total_score = 0
        articles = news_data['articles']
        for article in articles:
            sentiment_score = analyzer.polarity_scores(article['title'])
            total_score += sentiment_score['compound']
        avg_sentiment = total_score / len(articles) if articles else 0
        return avg_sentiment
    return 0

# Function to calculate RSI using yFinance
def fetch_technical_indicators(ticker):
    print(f"Fetching technical indicators for {ticker}")
    data = yf.download(ticker, period='1mo', interval='1d')
    data['RSI'] = (100 - (100 / (1 + (data['Close'].diff().clip(lower=0).rolling(window=14).mean() /
                                     abs(data['Close'].diff()).rolling(window=14).mean()))))
    rsi = data['RSI'].iloc[-1]
    return data, rsi

# Function to calculate Value at Risk (VaR)
def calculate_var(ticker, confidence_level=0.95):
    print(f"Calculating VaR for {ticker}")
    data = yf.download(ticker, period='3mo', interval='1d')
    returns = np.log(data['Close'] / data['Close'].shift(1)).dropna()
    var = np.percentile(returns, (1 - confidence_level) * 100)
    return var * np.sqrt(10)  # 10-day VaR

# Plot closing prices
def plot_stock_performance(stock_data, ticker):
    stock_data['Close'].plot(title=f"{ticker} - Closing Prices")
    plt.xlabel("Date")
    plt.ylabel("Close Price")
    plt.grid(True)
    plt.show()

# Plot RSI
def plot_rsi(stock_data, ticker):
    plt.figure(figsize=(10, 6))
    plt.plot(stock_data['RSI'], label='RSI')
    plt.axhline(70, color='red', linestyle='--', label='Overbought')
    plt.axhline(30, color='green', linestyle='--', label='Oversold')
    plt.title(f'{ticker} RSI')
    plt.legend()
    plt.show()

# Function to assess todays performance with summary and guidance
def assess_today_performance(date=None):
    # If no date is passed, use today's date
    if date is None:
        date = pd.Timestamp.today().date()
        date = pd.Timestamp.today().date() - timedelta(days=1)

    print(f"\nFetching activities for {date}...")

    # Fetch all activities from the Alpaca account
    activities = api.get_activities()

    pnl_summary = {
        'Ticker': [],
        'PnL': [],
        'RSI': [],
        'Sentiment': [],
        'VaR': []
    }

    # Dictionary to track bought positions
    position_tracker = {}

    for ticker in tickers:
        print(f"\n--- Assessing {ticker} ---")

        # Filter trades for the given ticker on the specified date
        orders_today = [
            o for o in activities 
            if (
                (pd.to_datetime(o.transaction_time).date() == date if hasattr(o, 'transaction_time') else pd.to_datetime(o.date).date() == date)
                and o.symbol == ticker 
                and o.activity_type == 'FILL'
            )
        ]

        if not orders_today:
            print(f"No trades for {ticker} on {date}.")
            pnl_summary['Ticker'].append(ticker)
            pnl_summary['PnL'].append(0)
            pnl_summary['RSI'].append(None)  # Placeholder for RSI
            pnl_summary['Sentiment'].append(None)  # Placeholder for sentiment
            pnl_summary['VaR'].append(None)  # Placeholder for VaR
            continue

        total_pnl = 0
        for order in orders_today:
            if order.side == 'buy':
                position_tracker[ticker] = float(order.price)
                print(f"Bought {order.qty} shares of {ticker} at {order.price}")
            elif order.side == 'sell' and ticker in position_tracker:
                buy_price = position_tracker[ticker]
                pnl = float(order.qty) * (float(order.price) - buy_price)
                total_pnl += pnl
                print(f"Sold {order.qty} shares of {ticker} at {order.price}, PnL: {pnl}")

        # Add to pnl_summary
        pnl_summary['Ticker'].append(ticker)
        pnl_summary['PnL'].append(total_pnl)
        pnl_summary['RSI'].append(None)  # Replace with actual RSI calculation
        pnl_summary['Sentiment'].append(None)  # Replace with actual sentiment calculation
        pnl_summary['VaR'].append(None)  # Replace with actual VaR calculation

    return pnl_summary

# Generate summary report with visual insights
def generate_summary_report(pnl_summary):
    # Convert pnl_summary to a DataFrame
    pnl_df = pd.DataFrame(pnl_summary)

    if not pnl_df.empty:
        print("\n--- Summary of Today's Performance ---")
        print(pnl_df)
        
        # Plot summary metrics
        fig, axs = plt.subplots(2, 2, figsize=(14, 10))

        sns.barplot(data=pnl_df, x='Ticker', y='PnL', ax=axs[0, 0])
        axs[0, 0].set_title("PnL per Stock")

        sns.barplot(data=pnl_df, x='Ticker', y='RSI', ax=axs[0, 1])
        axs[0, 1].axhline(70, color='red', linestyle='--', label='Overbought')
        axs[0, 1].axhline(30, color='green', linestyle='--', label='Oversold')
        axs[0, 1].set_title("RSI per Stock")

        sns.barplot(data=pnl_df, x='Ticker', y='Sentiment', ax=axs[1, 0])
        axs[1, 0].set_title("Sentiment per Stock")

        sns.barplot(data=pnl_df, x='Ticker', y='VaR', ax=axs[1, 1])
        axs[1, 1].set_title("VaR per Stock")

        plt.tight_layout()
        plt.show()

    else:
        print("No trades to summarize today.")

# Run the performance summary for today
if __name__ == "__main__":
    pnl_summary = assess_today_performance()
    generate_summary_report(pnl_summary)
```
